### Exploring the price and odometer columns

```python
autos["price"] = autos["price"].str.replace("$","")
autos["price"] = autos["price"].str.replace(",","")
autos["price"] = autos["price"].astype(float)
```

```python
autos["odometer"] = autos["odometer"].str.replace("km","")
autos["odometer"] = autos["odometer"].str.replace(",","")
autos["odometer"] = autos["odometer"].astype(float)
autos.rename({"odometer":"odometer_km"}, axis=1, inplace=True)
```

### Explore the brand column

```python
brand_groups = autos["brand"].value_counts(normalize=True)
common_brands = brand_groups[brand_groups > 0.05].index
print(common_brands)
```

In the cell above you can the the most frequent brands appearing in this dataset (Volkswagen, BMW, Opel, Mercedes Benz, Audi and Ford). They make up the groups of brands that are larger than 5%. It is noticable that the majority of these groups are brands from Germany.

```python
aggregate_dict_mean_prices = {}

for row in common_brands:
    selected_rows = autos[autos["brand"] == row]
    mean_price = selected_rows["price"].mean()
    aggregate_dict_mean_prices[row] = mean_price

aggregate_dict_mean_prices
```

The most frequent brands in the data set are ranging in mean prices from 2975 $ (Opel) to 9336 $ (Audi). The most frequent brand Volkswagen shows a mean price of 5402 $. Overall Audi, BMW and Mercedes are more expensive, Ford and Open are less expensive and Volkswagwn is in between.

```python
mean_prices_series = pd.Series(aggregate_dict_mean_prices)
print(mean_prices_series)

df = pd.DataFrame(mean_prices_series, columns = ["mean_price"])
print(df)
```

```python
brand_mean_mileage = {}

for row in common_brands:
    selected_rows = autos[autos["brand"] == row]
    mean_odometer = selected_rows["odometer_km"].mean()
    brand_mean_mileage[row] = int(mean_odometer)
    
brand_mean_mileage
```

```python
brand_mean_mileage = pd.Series(brand_mean_mileage)
print(brand_mean_mileage)
```

```python
brand_info = pd.DataFrame(brand_mean_mileage, columns = ["mean_mileage"])
print(brand_info)
```

```python
brand_info["mean_price"] = mean_prices_series
brand_info
```

Even though the mean price in ranging largely for the top brands, the mean_mileage for Volkswagen, BMW, Opel, Mercedes Benz, Audi and Ford remains similar, ranging in between 120.000 and 130.000.

There is a slight trend with the more expensive vehicles having a lower mileage, versus the more affordable vehicles having a higher mileage.
